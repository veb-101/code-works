```python
from scipy.stats import norm
```

```python
norm.pdf(0) # probablity density of 0 from standard normal distribution
```

```python
norm.pdf(0, loc=5, scale=10) # Gaussioan distribution loc=mean, scale = std deviation
```

```python
import numpy as np
r = np.random.randn(10)
norm.pdf(r)
```

```python
norm.logpdf(r)
```

```python
norm.cdf(r) # cumulative distributive function
```

```python
norm.logcdf(r)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fourier transform converts a signal from time domain to frequency domain
x = np.linspace(0,100, 10000)
y = np.sin(x) + np.sin(3*x) + np.sin(5*x)
plt.plot(y)
plt.show()
```

```python
Y = np.fft.fft(y) # gives signal with complex numbers so find mag. before plotting
plt.plot(np.abs(Y))
plt.show()
```
